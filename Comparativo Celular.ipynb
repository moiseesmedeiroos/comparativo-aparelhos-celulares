```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import warnings
```

```python
# Desconsiderando warnings
warnings.filterwarnings("ignore")
```

```python
# Upando dataset
comparativo_cell = 'comparativo celulares.csv'
df_cell = pd.read_csv(comparativo_cell, sep=";")
```

```python
df_cell
```

```python
# Removendo linhas NaN
index_remover = [7,8,9]
```

```python
df_cell = df_cell.drop(index_remover)
```

```python
df_cell
```

```python
df_cell.dtypes
```

```python
# Alterando o tipo dos valores de algumas colunas
df_cell[['Custo-Benefício','Hardware','Qualidade da Câmera','Qualidade da Tela','Desempenho']] = df_cell[['Custo-Benefício','Hardware','Qualidade da Câmera','Qualidade da Tela','Desempenho']].astype(float)
df_cell.dtypes
```

Como todos os aparelhoes têm **5g, Dual Sim, NFC(Tecnologia de aproximação) e Meméria RAM de 8gb**, excluiremos as respectivas colunas.

```python
# Excluindo colunas
df_cell = df_cell.drop(['Dual Sim?','5g?','RAM','NFC(Tecnologia de aproximação)'], axis=1)
```

```python
df_cell
```

# Comparativo dos aparelhos segundo notas da crítica especializada:

```python
# Selecionando colunas com notas da crítica
df_cb= df_cell[['Modelo','Custo-Benefício','Hardware','Qualidade da Tela','Qualidade da Câmera','Desempenho']]
df_cb
```

#### Vamos analisar graficamente:

```python
# Configurações do Seaborn
sns.set(style="whitegrid")

# Configurações do matplotlib
plt.figure(figsize=(20,6))


# Custo-Benefício
plt.subplot(1, 3, 1)
sns.barplot(x='Modelo', y='Custo-Benefício', data=df_cb, palette='viridis')
plt.title('Custo-Benefício')
plt.xticks(rotation = 40)

plt.ylabel('') # Remove legenda y
plt.xlabel('') # Remove legenda x


# Hardware
plt.subplot(1, 3, 2)
sns.barplot(x='Modelo', y='Hardware', data=df_cb, palette='viridis')
plt.title('Hardware')
plt.xticks(rotation = 40)

plt.ylabel('') # Remove legenda y


# Qualidade da Tela
plt.subplot(1, 3, 3)
sns.barplot(x='Modelo', y='Qualidade da Tela', data=df_cb, palette='viridis')
plt.title('Qualidade da Tela')
plt.xticks(rotation = 40)

plt.ylabel('') # Remove legenda y
plt.xlabel('') # Remove legenda x


# Ajustes de layout
plt.tight_layout()


# Exibindo os gráficos
plt.show()
```

```python
# Configurações do Seaborn
sns.set(style="whitegrid")

# Configurações do matplotlib
plt.figure(figsize=(25, 6))


# Qualidade da Câmera
plt.subplot(1, 5, 4)
sns.barplot(x='Modelo', y='Qualidade da Câmera', data=df_cb, palette='viridis')
plt.title('Qualidade da Câmera')
plt.xticks(rotation = 40)

plt.ylabel('')


# Desempenho
plt.subplot(1, 5, 5)
sns.barplot(x='Modelo', y='Desempenho', data=df_cb, palette='viridis')
plt.title('Desempenho')
plt.xticks(rotation = 40)

plt.ylabel('')


# Ajustes de layout
plt.tight_layout()


# Exibindo os gráficos
plt.show()
```
